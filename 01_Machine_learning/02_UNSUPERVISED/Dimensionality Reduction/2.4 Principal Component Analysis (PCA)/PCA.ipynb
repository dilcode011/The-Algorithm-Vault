```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer_data = load_breast_cancer()
df = pd.DataFrame(cancer_data['data'], columns=cancer_data['feature_names'])
```

```python
print(f"Original Data Shape: {df.shape}")
# Observation: 569 patients, 30 different features. We cannot graph 30 dimensions!
```

```python
# PCA calculates variance. If we don't scale, features with larger numbers 
# (like 'mean area' which is in the 1000s) will dominate the PCA.

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# Let's compress 30 dimensions down to 2 dimensions.
pca = PCA(n_components=2)
x_pca = pca.fit_transform(scaled_data)

print(f"Compressed Data Shape: {x_pca.shape}")
# Result: 569 patients, but now only 2 features (Principal Component 1 & 2)
```

```python
plt.figure(figsize=(10, 6))

# Plot the two new components. Color them by their actual diagnosis (Malignant vs Benign)
scatter = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer_data['target'], cmap='plasma', alpha=0.7)

plt.title('PCA: 30-Dimension Cancer Data Compressed to 2D')
plt.xlabel('First Principal Component (PC1)')
plt.ylabel('Second Principal Component (PC2)')
plt.legend(handles=scatter.legend_elements(), labels=['Malignant', 'Benign'])
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# OBSERVATION:
# Even though we threw away 28 columns of data, PCA kept enough of the "core essence" 
# that we can clearly see two distinct clusters of Malignant and Benign tumors!
```

```python
# How much information did we actually keep?
variance = pca.explained_variance_ratio_

print("\n--- Information Retained ---")
print(f"PC1 holds {variance*100:.2f}% of the original data's information.")
print(f"PC2 holds {variance*100:.2f}% of the original data's information.")
print(f"Total information kept: {(variance + variance)*100:.2f}%")

# Even with just 2 components, we usually keep over 60% of the variance from all 30 features.
```
