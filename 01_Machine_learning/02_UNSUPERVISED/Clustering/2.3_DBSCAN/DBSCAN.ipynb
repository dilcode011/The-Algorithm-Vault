```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# We create data that is shaped like two interlocking half-moons.
# We also add some noise (random scattered points).
X, y = make_moons(n_samples=400, noise=0.1, random_state=42)
```

```python
# Standardizing data is highly recommended for DBSCAN
# so that the 'eps' distance metric works uniformly.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='gray', s=30)
plt.title("Raw Unlabeled Data (Interlocking Moons + Noise)")
plt.show()
```

```python
# Let's see what happens if we use K-Means (assuming K=2)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=200, c='red', marker='X', label='Centroids')
plt.title("K-Means Attempt (Fails to capture the moon shapes)")
plt.legend()
plt.show()
```

```python
# OBSERVATION:
# K-Means just draws a straight line down the middle. 
# It assumes clusters are circles, so it cuts the moons in half.
```

```python
# We don't tell it how many clusters there are!
# We only give it the search radius (eps) and minimum points (min_samples)
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
# Note: DBSCAN assigns a label of '-1' to points it considers Noise/Outliers
scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='plasma', s=30)
plt.title("DBSCAN Attempt (Successfully isolates moons & detects noise)")
plt.colorbar(scatter, label='Cluster Label (-1 = Noise)')
plt.show()

```

```python
# OBSERVATION:
# 1. It perfectly separated the two moons (e.g., Yellow and Purple).
# 2. It identified the scattered points as Outliers and labeled them -1 (Dark Blue).
```
