```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# We'll create 5 distinct groups of data points
X, y_true = make_blobs(n_samples=200, centers=5, cluster_std=0.8, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, color='gray')
plt.title("Raw Unlabeled Data")
plt.show()
```

```python
# We use the dendrogram to figure out how many clusters we actually have.
# 'ward' linkage minimizes the variance of the clusters being merged.

plt.figure(figsize=(12, 8))
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points (Index)')
plt.ylabel('Euclidean Distances')

# Generate the linkage matrix and plot
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
# OBSERVATION:
# To find the optimal number of clusters, find the longest vertical line 
# that is NOT intersected by any horizontal line.
# If we draw a horizontal line across that longest vertical gap, 
# it intersects 5 vertical lines. Therefore, optimal K = 5.
plt.axhline(y=15, color='r', linestyle='--') # Example cut line
plt.show()
```

```python
# Now that the dendrogram showed us K=5 is optimal, we build the model.

hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.figure(figsize=(10, 8))

# Dynamically plot however many clusters we chose
colors = ['red', 'blue', 'green', 'purple', 'orange']
for i in range(5):
    plt.scatter(X[y_hc == i, 0], X[y_hc == i, 1], s=50, c=colors[i], label=f'Cluster {i+1}')

plt.title('Hierarchical Clustering Results (Agglomerative)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
